# Interpretation

# Top products

```python
top_products = df[['title','boughtInLastMonth','stars','price']]\
.sort_values(by='boughtInLastMonth',ascending=False)\
.head(10)
top_products
```

# Top 10 most purchased products

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data = top_products,
    x = 'boughtInLastMonth',
    y = 'title'
)
plt.title('Top 10 Most Purchased Products')
plt.xlabel('Purchases in last month')
plt.ylabel('Product')
plt.show()
```

# Creating short version of product purchaseed for clearer understanding

```python
top_products = data[['title','boughtInLastMonth','stars','price']]\
.sort_values(by='boughtInLastMonth',ascending=False)\
.head(10)
top_products['short_title'] = top_products['title'].str[:50] + '...'
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data=top_products,
    x='boughtInLastMonth',
    y='short_title'
)
plt.title('Top 10 most purchased products')
plt.xlabel('purchase in last month')
plt.ylabel('Product')
plt.show()
```

# Correlation heatmap

```python
corr = data[['stars','reviews','price','boughtInLastMonth']].corr()
plt.figure(figsize=(8,5))
sns.heatmap(corr,annot=True,cmap='Blues')
plt.title('Correlation Between Product Metrics')
plt.show()
```

# Top ten categories by number of products

```python
top_categories = (
    data['category_id']
    .value_counts()
    .head(10)
)
plt.figure(figsize=(10,6))
sns.barplot(
    x = top_categories.values,
    y = top_categories.index
)
plt.title('Top 10 Categories by Number of Products')
plt.xlabel('Number of Products')
plt.ylabel('Categories ID')
plt.show()
```

# Key Findings
